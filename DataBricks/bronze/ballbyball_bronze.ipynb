```python
path = r'abfss://bronze-external@ipldatastorage.dfs.core.windows.net/Matches'
filename = r'1_Match_id_335987.parquet'
```

```python
df = spark.read.format('parquet').load(path+filename)
```

```python
from pyspark.sql.functions import current_timestamp, input_file_name

df = df.withColumn('insert_ts', current_timestamp())
df = df.withColumn('file_name', input_file_name())
```

```python
from pyspark.sql.functions import concat_ws

df =  df.withColumn('row_id', concat_ws('-', 'Match_id', 'Innings_No', 'Over_id', 'Ball_id'))
```

```python
df.write.format('delta')\
  .mode('append')\
  .saveAsTable('bronze.ball_by_ball')
```
