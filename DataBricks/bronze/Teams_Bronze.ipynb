```python
path = r'abfss://bronze-external@ipldatastorage.dfs.core.windows.net/Team/'
filename = r'Team.parquet'
```

```python
df = spark.read.format('parquet').load(path+filename)
```

```python
from pyspark.sql.functions import current_timestamp, lit

df = df.withColumn('insert_ts', current_timestamp())
df = df.withColumn('file_name', lit(filename))
```

```python
df.columns
```

```python
df.write.format('delta')\
  .mode('append')\
  .saveAsTable('ipl.bronze.team')
```
